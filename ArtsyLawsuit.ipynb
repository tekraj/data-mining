```python
import pandas as pd
```

```python
artsy_data_file = "./dataset/artsy_lawsuit.xlsx"
artsy_data = pd.read_excel(artsy_data_file,sheet_name=1)
artsy_data.head()
```

```python
import numpy as np
# average rate
average_rate = artsy_data.groupby('Gender').mean()
average_rate
```

```python
# standard deviataion
standard_deviation = artsy_data.groupby('Gender').describe()
standard_deviation
```
